````python
def create_features_updated(data, max_lag, rolling_mean_size):
    data['day_of_week'] = data.index.dayofweek
    data['hour'] = data.index.hour
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
    for lag in range(1, max_lag + 1):
        data['week_before_lag_{}'.format(lag)] = data['num_orders'].shift(lag + 168)    
    data['rolling_mean'] = data['num_orders'].shift()
    data['rolling_mean'] = data['rolling_mean'].rolling(rolling_mean_size).mean()
````

### –ú–µ—Ç–æ–¥ split_data

````python
def split_data(data, column_name):
    # 1st we separate data for features and target data
    features = data.loc[:,data.columns != column_name]
    target = data[column_name]
    # 2rd we separate data for testing and training samples
    features_train, features_test, target_train, target_test =\
    train_test_split(
            features,
            target,
            train_size=0.9,
            shuffle=False
    )
    return features_train, features_test, target_train, target_test
````

````python
features_train, features_test, target_train, target_test =  split_data(data, 'num_orders')
````

### –ú–µ—Ç–æ–¥ lgm_teach

````python
def lgm_teach(features_train, target_train, cv):
    param_grid = {
        'num_leaves': [31, 127],
        'boosting_type': ['gbdt'],
        'metric': ['rmse'],
        'colsample_bytree': [0.9],
        'objective': ['rmse'],
        'subsample': [0.8],
        'learning_rate': [0.05],
        'subsample_freq': [5],
        'verbose': [-1],
        'n_estimators': [100, 200],
        'min_child_samples': [30, 50, 100]
        }

    lgb_estimator = lgb.LGBMRegressor(random_state=RANDOM_STATE)

    gsearch = GridSearchCV(estimator=lgb_estimator, param_grid=param_grid, cv=cv, scoring='neg_mean_squared_error')
    lgb_model = gsearch.fit(features_train, target_train)
    return lgb_model
````
