#### –ú–µ—Ç–æ–¥ data_split

```python
def data_split(data, decypher=False, altered = False, A = None, a = None):
    # 1st we separate data for features and target data
    features = data.drop(data_drop_index)[data.columns[data.columns != 'insurance_count']]
    target = data.drop(data_drop_index)['insurance_count']
    # 2rd we separate data for testing and training samples
    features_train, features_test, target_train, target_test =\
    train_test_split(
        features,
        target,
        train_size=0.7,
        random_state=12345,
        stratify=target
    )
    if altered:
        if decypher:
           # A = A.loc[features_train.index, features_train.index]
            features_train = (np.linalg.inv(A.loc[features_train.index, features_train.index]) @ features_train).set_index(features_train.index)
            features_train = (A.loc[features_train.index, features_train.index].T @ features_train).set_index(features_train.index)
        else:
            features_train = features_train.dot(A)
            features_test = features_test.dot(A)
    return features_train, features_test, target_train, target_test, features
```

```python
features_train, features_test, target_train, target_test, _ = data_split(data)
```

#### –ö–ª–∞—Å—Å OtherLinearRegression

```python
class OtherLinearRegression:
    def fit(self, train_features, train_target):
        X = np.concatenate((np.ones((train_features.shape[0], 1)), train_features), axis=1)
        y = train_target
        w = np.linalg.inv(X.T.dot(X)) @ X.T @ y 
        self.w = w[1:]
        self.w0 = w[0]

    def predict(self, test_features):
        return test_features.dot(self.w) + self.w0
    
```

```python
model = OtherLinearRegression()
model.fit(features_train, target_train)
predictions = model.predict(features_test)
predicted_score = mean_squared_error(target_test, predictions)
predicted_r2_score = r2_score(target_test, predictions)
```

#### –ú–µ—Ç–æ–¥ dummy_check

```python
def dummy_check(features_train,  target_train,features_test, target_test, predicted_score):
    dummy_regr = DummyRegressor(strategy="mean")
    dummy_regr.fit(features_train, target_train)
    dummy_score = mean_squared_error(dummy_regr.predict(target_test), target_test)
    print(f'Dummy predicted - {dummy_score:.3f}, model predicted - {predicted_score:.3f}')
    if dummy_score  > predicted_score:
        print('Model works fine, still better than dummy')
    else:
        print('Model is not very helpful')
    return mean_squared_error(dummy_regr.predict(target_test), target_test)
```

#### –ú–µ—Ç–æ–¥ generate_invertible_matrix

```python
def generate_invertible_matrix(n, rng):
    matrix = rng.rand(n,n)
    while True:
        try:
            np.linalg.inv(matrix)
            break
        except:
            matrix = rng.rand(n,n)
    return matrix 
```

```python
rng = np.random.RandomState(12345)
```

```python
def altered_check(data, predicted_r2_score, model, decypher = False):
    # "A" matrix to protect features
   
    A = generate_invertible_matrix(data.shape[0],rng)
    A = pd.DataFrame(A)
    features_train, features_test, target_train, target_test, features = data_split(data, decypher,  True, A)
    if decypher:
        features_train = (np.linalg.inv(A.loc[features_train.index, features_train.index].T) @ features_train).set_index(features_train.index)
        features_train = (A.loc[features_train.index, features_train.index]  @ features_train).set_index(features_train.index)
    model.fit(features_train, target_train)
    altered_predicted_r2_score = r2_score(target_test, model.predict(features_test))
    print(f'Altered predicted r2 score - {altered_predicted_r2_score:.3f}, model predicted - {predicted_r2_score:.3f}')
    return A
```
